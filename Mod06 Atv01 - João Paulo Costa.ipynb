#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 

sinasc = pd.read_csv('sinasc_RO_2019.csv')
sinasc.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.rc('figure', figsize=(20, 10))
fig, axes = plt.subplots(2, 3)

sns.countplot(ax = axes[0, 0], x='APGAR1', data=sinasc)
sns.countplot(ax = axes[0, 1], x='APGAR5', data=sinasc)
sns.countplot(ax = axes[0, 2], x='SEXO', data=sinasc)
sns.countplot(ax = axes[1, 0], x='GESTACAO', data=sinasc).tick_params(axis="x", rotation = 35)
sns.countplot(ax = axes[1, 1], x='ESTCIVMAE', data=sinasc)
plt.tight_layout()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
ap5_por_cons = sinasc[['CONSULTAS', 'APGAR5']].dropna()

plt.figure(figsize=(8, 5))
sns.barplot(data=ap5_por_cons, x='CONSULTAS', y='APGAR5', hue='CONSULTAS', palette='crest', legend=False)
plt.title('APGAR5 médio por número de consultas pré-natais')
plt.xlabel('Número de consultas pré-natais')
plt.ylabel('APGAR5 médio')
plt.tight_layout()
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(10, 7))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0, 0], x='SEXO', data=sinasc)
sns.countplot(ax = axes[0, 1], x='GESTACAO', data=sinasc).tick_params(axis="x", rotation = 35)
sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=sinasc)
sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=sinasc)
plt.tight_layout()
plt.show()
```
